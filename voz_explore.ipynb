```python
from bs4 import BeautifulSoup
import requests
import os
import re
```

```python
def get_soup(url):
    html = requests.get(url).text
    soup = BeautifulSoup(html, "html.parser")
    return soup
```

```python
homepage_url = "https://voz.vn"
homepage_soup = get_soup(homepage_url)

topics = []
titles = homepage_soup.find_all("h3", class_="node-title")
for t in titles:
    links = [f"{homepage_url}{x['href']}" for x in t.find_all("a")]
    topics.extend(links)

topics
```

```python
threads = []
count = 0

url = topics[4]
print(url)
while True:
    count += 1
    topic_soup = get_soup(url)

    links = topic_soup.find_all("a", attrs={"data-tp-primary": "on"})
    links = [f"{homepage_url}{x['href']}" for x in links]
    threads.extend(links)
    
    next_button = topic_soup.find("a", class_="pageNav-jump--next")
    if next_button and count < 5:
        url = f"{homepage_url}{next_button['href']}"
    else:
        break
    
threads
```

```python
# url = "https://voz.vn/t/thac-mac-ly-do-bi-ban-post-vao-day-khong-lap-thread-ngoai-doc-ky-post-1.588/page-20"
url = "https://voz.vn/t/vozer-sang-tac-hoa-am-choi-nhac.3254/"
posts = []
count = 0

def process_post(p):
    for script in p.find_all("script"):
        script.decompose()
    for reply in p.find_all("blockquote"):
        reply.decompose()
    for link in p.find_all("a"):
        link.decompose()

    p = p.find_all(text=True)
    p = " ".join(p).strip()
    p = re.sub(r"\n+", " ", p)
    p = re.sub(r"\s+", " ", p)

    return p

while True:
    count += 1
    soup = get_soup(url)
    print(url)
    post_containers = soup.find_all("div", class_="bbWrapper")

    for p in post_containers:
        p = process_post(p)
        if p:
            posts.append(p)
            
    next_button = soup.find("a", class_="pageNav-jump--next")
    if next_button and count < 5:
        url = f"{homepage_url}{next_button['href']}"
    else:
        break

posts
```
