```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import os
from selenium import webdriver

geckodriver_path = "C:\\Users\\Thien\\Documents\\GitHub\\VN-text-datasets\\geckodriver.exe"
# current_dir = os.getcwd()
```

```python
url = "https://voz.vn/"

with webdriver.Firefox(executable_path=geckodriver_path) as driver:
    driver.get(url)
    topics = [x.get_attribute("href") for x in driver.find_elements_by_css_selector(".node-title > a")]

topics
```

```python
with webdriver.Firefox(executable_path=geckodriver_path) as driver:
    driver.get(topics[6])
    rooms = [x.get_attribute("href") for x in driver.find_elements_by_css_selector(".structItem-title > a")]
    
    # go to the next page
    count = 1
    max_count = 5
    try:
        next_button = driver.find_element_by_css_selector(".pageNav-jump--next")
    except:
        next_button = None
    while next_button and count < max_count:
        next_button.click()
        count += 1
        rooms.extend([x.get_attribute("href") for x in driver.find_elements_by_css_selector(".structItem-title > a")])
        try:
            next_button = driver.find_element_by_css_selector(".pageNav-jump--next")
        except:
            next_button = None
    
rooms
```

```python
with webdriver.Firefox(executable_path=geckodriver_path) as driver:
    # driver.get(rooms[0])
    driver.get("https://voz.vn/t/vozer-sang-tac-hoa-am-choi-nhac.3254/")
    texts = []
    convo = [BeautifulSoup(x.get_property("outerHTML")).div.findAll(text=True, recursive=False) for x in driver.find_elements_by_css_selector("div.bbWrapper")]
    convo = [[x.strip() for x in text if x.strip()] for text in convo]
    texts.extend(convo)
    
    # go to next page
    count = 1
    max_count = 10
    try:
        next_button = driver.find_element_by_css_selector(".pageNav-jump--next")
    except:
        next_button = None
    while next_button and count < max_count:
        next_button.click()
        count += 1
        convo = [BeautifulSoup(x.get_property("outerHTML")).div.findAll(text=True, recursive=False) for x in driver.find_elements_by_css_selector("div.bbWrapper")]
        convo = [[x.strip() for x in text if x.strip()] for text in convo]
        texts.extend(convo)
        try:
            next_button = driver.find_element_by_css_selector(".pageNav-jump--next")
        except:
            next_button = None
    
texts
```
